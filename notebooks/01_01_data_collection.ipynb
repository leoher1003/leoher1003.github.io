```python
import pandas as pd
import time
import os
from tqdm import tqdm
from pybaseball import batting_stats, pitching_stats

# Create directory if it doesn't exist
os.makedirs('../data/raw_data', exist_ok=True)

years = range(2010, 2026) # 16 years of data from 2010 to 2025
all_batters = []
all_pitchers = []

print("[INFO] Downloading Season Data (2011-2024)...")

for year in tqdm(range(2011, 2025), desc="Scraping Progress"):
    print(f"[{year}] Collecting batter data...")
    # qual=0 : Collect all batters with at least 1 plate appearance
    bat = batting_stats(year, qual=0)
    bat['Season'] = year
    bat['Role'] = 'Batter'
    
    # Keep only essential columns to prevent RAM overload
    # (Season, Name, Team, WAR, Role, Games, Plate Appearances)
    bat = bat[['Season', 'Name', 'Team', 'WAR', 'Role', 'G', 'PA']]
    all_batters.append(bat)
    
    # Delay to prevent server ban (DO NOT REMOVE)
    time.sleep(3) 
    
    print(f"[{year}] Collecting pitcher data...")
    # qual=0 : Collect all pitchers who threw at least 1 pitch
    pit = pitching_stats(year, qual=0)
    pit['Season'] = year
    pit['Role'] = 'Pitcher'
    
    # Keep essential columns for pitchers
    # (G, GS, IP are crucial later to classify starters vs. relievers)
    pit = pit[['Season', 'Name', 'Team', 'WAR', 'Role', 'G', 'GS', 'IP']]
    all_pitchers.append(pit)
    
    # Another delay
    time.sleep(3)

# Merge the lists into DataFrames
df_batters = pd.concat(all_batters, ignore_index=True)
df_pitchers = pd.concat(all_pitchers, ignore_index=True)

# Save to local CSV (Will load these in 02_feature_engineering for preprocessing)
batters_path = '../data/raw_data/raw_batters_2010_2025.csv'
pitchers_path = '../data/raw_data/raw_pitchers_2010_2025.csv'

df_batters.to_csv(batters_path, index=False)
df_pitchers.to_csv(pitchers_path, index=False)

print(f"Collection complete!")
print(f"Batter data: {len(df_batters)} rows saved to {batters_path}")
print(f"Pitcher data: {len(df_pitchers)} rows saved to {pitchers_path}")
```
